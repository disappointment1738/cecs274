## RUN-TIME

1. (1 pt each) Determine the big-O run-time for each algorithm below.

(a) 
`def clone(array):
""" returns a copy of the given array """
    copy = []
    for e in array: # iterates e times (length of array)
        copy.append(e) # 1 visit per iteration
    return copy`

Answer:
`for e in array` will iterate `e` times, `e` being the length of the array.
`copy.append(e)` will be visited 1 time for every iteration.
the total # of visits is `e*1`
since `e` is basically `n`, f(n) = n
by definition, g(n) = O(f(n))
therefore, the run-time is O(n)


(b)
`def integer_div(a, b):
"""performs integer division a // b """
    count = 0
    total = b
    while total <= a: # iterates *** times
        total += b # 1 visit per iteration
        count += 1 # 1 visit per iteration
    return count`

Answer: 
`while total <= a` will iterate **** times.
inside the `while` loop, there are 2 visits per iteration.
the total # of visits is a certain number called `a`, the number being smaller than a, multipled by `2` and f(n) = 2a
by definition, g(n) = O(f(n))
therefore, the run-time is O(**), or O(1)

Further reading (https://inst.eecs.berkeley.edu/~cs10/labs/cur/programming/algorithms/timing/constant-time.html?topic=berkeley_bjc%2Fareas%2Falgorithm-complexity.topic) and (https://hackernoon.com/big-o-for-beginners-622a64760e2) and (https://en.wikipedia.org/wiki/Computational_complexity_of_mathematical_operations) and (https://www.cse.iitk.ac.in/users/manindra/CS681/lecture1and2.pdf)

(c) 
`def mod(a : int, b : int):
"""returns the remainder of a divided by b"""
    if b <= 0: # 1 visit
        return -100
    quotient = a // b # 1 visit
    return a - quotient * b`

Answer: 
`if b <= 0` will be visited 1 time
`quotient = a // b` will be visited 1 time
the total # of visits is `1+1` = `2` and f(n) = 2
by definition, g(n) = O(f(n))
therefore, the run-time is O(2), or O(1)

- `O(N + log N)`
- `O(N + N log N)`
